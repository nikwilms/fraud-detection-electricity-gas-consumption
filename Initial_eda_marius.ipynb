```python
# number of transactions per client by adding the number of unique client_id's per client

df_combined['number_of_transactions'] = df_combined.groupby('client_id')['client_id'].transform('count')

# plot number of transactions per client vs target in a boxplot

plt.figure(figsize=(10,10))
sns.boxplot(x='target', y='number_of_transactions', data=df_combined)
plt.title('number of transactions per client vs target')
plt.show()

```

```python
# relationship between target and tarif type

tarif_type = df_combined.groupby(['tarif_type', 'target'])['client_id'].count().reset_index()
tarif_type.columns = ['tarif_type', 'target', 'count']
tarif_type = tarif_type.pivot(index='tarif_type', columns='target', values='count')
tarif_type['total'] = tarif_type[0] + tarif_type[1]
tarif_type['no_fraud'] = tarif_type[0] / tarif_type['total']
tarif_type['fraud'] = tarif_type[1] / tarif_type['total']
tarif_type = tarif_type.sort_values(by='no_fraud')
tarif_type[['no_fraud', 'fraud']].plot.bar(stacked=True)
plt.title('Tarif type distribution')
plt.show()
```

```python
# see if creation date is related to target

df_combined['creation_date'] = pd.to_datetime(df_combined['creation_date'])
df_combined.sort_values(by='creation_date', inplace=True)

# bin creation dates into yearly bins

bin_list = list(range(1980, 2020,1))

df_combined['creation_date'] = pd.cut(df_combined['creation_date'].dt.year, bins=40, labels=bin_list)

# plot creation date vs target

creation_date = df_combined.groupby(['creation_date', 'target'])['client_id'].count().reset_index()
creation_date.columns = ['creation_date', 'target', 'count']
creation_date = creation_date.pivot(index='creation_date', columns='target', values='count')
creation_date['total'] = creation_date[0] + creation_date[1]
creation_date['no_fraud'] = creation_date[0] / creation_date['total']
creation_date['fraud'] = creation_date[1] / creation_date['total']
creation_date = creation_date.sort_values(by='no_fraud')
creation_date[['no_fraud', 'fraud']].plot.bar(stacked=True)
plt.title('creation date distribution')
plt.show()
```
